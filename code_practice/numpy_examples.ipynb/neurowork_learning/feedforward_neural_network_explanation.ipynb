# 前馈神经网络（Feedforward Neural Network）详解

## 什么是前馈神经网络？
前馈神经网络是最基本的人工神经网络类型之一。"前馈"这个术语来源于网络中信息的流动方向：数据只能从输入层向前流动，经过隐藏层，最后到达输出层，不存在循环或反向连接。

### 主要特点：
1. **单向传播**：信息只能从前往后传递
2. **层次结构**：包含输入层、隐藏层和输出层
3. **无循环**：节点之间的连接不形成环
4. **全连接**：每一层的节点都与下一层的所有节点相连

## 网络结构示意图

```python
import numpy as np
import matplotlib.pyplot as plt

def create_network_visualization():
    plt.figure(figsize=(10, 6))

    input_layer = np.array([[0, i] for i in range(3)])
    hidden_layer = np.array([[0.5, i] for i in range(4)])
    output_layer = np.array([[1, i] for i in range(2)])

    plt.scatter(input_layer[:,0], input_layer[:,1], c='blue', s=100, label='输入层')
    plt.scatter(hidden_layer[:,0], hidden_layer[:,1], c='green', s=100, label='隐藏层')
    plt.scatter(output_layer[:,0], output_layer[:,1], c='red', s=100, label='输出层')

    for i in input_layer:
        for h in hidden_layer:
            plt.plot([i[0], h[0]], [i[1], h[1]], 'blue', alpha=0.2)

    for h in hidden_layer:
        for o in output_layer:
            plt.plot([h[0], o[0]], [h[1], o[1]], 'red', alpha=0.2)

    plt.title('前馈神经网络结构示意图')
    plt.legend()
    plt.grid(True)
    plt.show()

create_network_visualization()
```

## 前馈神经网络的工作原理

### 1. 前向传播过程：
1. **输入层接收数据**：
   - 原始数据进入网络
   - 每个输入节点代表一个特征

2. **隐藏层处理**：
   - 接收来自输入层的加权信号
   - 通过激活函数处理信号
   - 可以有多个隐藏层

3. **输出层生成结果**：
   - 处理最后一个隐藏层的信号
   - 生成最终预测结果

### 2. 数学原理
每个神经元的输出计算如下：
1. 计算加权和：sum = Σ(输入值 × 权重) + 偏置
2. 将加权和传入激活函数：output = activation(sum)

```python
def neuron_demo(inputs, weights, bias):
    weighted_sum = np.dot(inputs, weights) + bias

    def sigmoid(x):
        return 1 / (1 + np.exp(-x))

    output = sigmoid(weighted_sum)
    return output

inputs = np.array([0.5, 0.3, 0.2])
weights = np.array([0.4, 0.6, 0.2])
bias = 0.1

result = neuron_demo(inputs, weights, bias)
print(f"神经元输入: {inputs}")
print(f"神经元权重: {weights}")
print(f"神经元偏置: {bias}")
print(f"神经元输出: {result:.4}")
```

## 前馈神经网络的应用场景

前馈神经网络适用于多种任务，包括：
1. **分类问题**：
   - 图像识别
   - 垃圾邮件检测
   - 客户分类

2. **回归问题**：
   - 房价预测
   - 销售预测
   - 温度预测

3. **特征学习**：
   - 数据降维
   - 特征提取

### 优点：
- 结构简单，易于理解
- 计算效率高
- 适用于多种问题类型

### 局限性：
- 不适合处理序列数据
- 不能捕捉时间依赖关系
- 需要固定大小的输入
